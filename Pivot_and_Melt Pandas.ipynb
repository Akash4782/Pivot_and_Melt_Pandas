```python
import pandas as pd 
var = pd.DataFrame({"Days":[1,2,3,4,5,6,7],"Eng":[11,12,13,14,15,16,17],"Maths":[21,22,23,24,25,26,27]})
var
```

```python
pd.melt(var) 
```

```python
pd.melt(var,id_vars = ["Eng"]) 
```

```python
pd.melt(var,id_vars = ["Days"]) 
```

```python
pd.melt(var,id_vars = ["Eng"] , var_name ="Python") 
```

```python
pd.melt(var,id_vars = ["Eng"] , var_name ="Python",value_name = "Numbers") 
```

```python
var = pd.DataFrame({"Days":[1,2,3,4,5,6,7],"Stu_name":["a","b","c","a","b","c","a"],"Eng":[11,12,13,14,15,16,17],"Maths":[21,22,23,24,25,26,27]})
var
```

```python
var.pivot(index = "Days" , columns = "Stu_name")
```

```python
var = pd.DataFrame({"Days":[1,2,3,4,5,6,7],"Stu_name":["a","b","a","a","b","b","a"],"Eng":[11,12,13,14,15,16,17],"Maths":[21,22,23,24,25,26,27]})
var
```

```python
var.pivot(index = "Days" , columns = "Stu_name")
```

```python
var.pivot(index = "Days" , columns = "Stu_name" , values = "Maths")
```

```python
var.pivot(index = "Days" , columns = "Stu_name" , values = "Eng")
```

```python
var = pd.DataFrame({"Days":[1,1,1,1,1,2,2],"Stu_name":["a","b","c","a","b","c","a"],"Eng":[11,12,13,14,15,16,17],"Maths":[21,22,23,24,25,26,27]})
var
```

```python
var.pivot_table(index = "Days" , columns = "Stu_name" , aggfunc = "mean")
```

```python
var.pivot_table(index = "Days" , columns = "Stu_name" , aggfunc = "sum")
```

```python
var.pivot_table(index = "Days" , columns = "Stu_name" , aggfunc = "sum",margins = "True")    # Margin shows an Average value
```

```python
var.pivot_table(index = "Days" , columns = "Stu_name" , aggfunc = "mean",margins = "True")
```
